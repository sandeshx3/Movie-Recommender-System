# Types of recommendation system
  1. content based : based on the content similarities, tags are created then those tags are compared.
  2. collaborative filtiring based : based on the user interest, suppose user A and user B , their similarities score is strong suppose 0.9, suppose userA like movie m2 based on their similiraties score is high , user B also get that movie recommendation
  3. hybrid: mix of content based and collabrative filtering

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

# now we need to calculate the distance as as smaller the distance the more similar the movie would be 
# here we don't calculate the eculudian distance,here we calculate the cosine distance
# simply eculudian distance for two point A(x1,y1) and B(x2,y2) is  sqrt(x2-x1)square+sqrt(y2-y1)square
# instead angle will be calculated also know as consine distance 
# higher the dimensional eculadian distance will fails
# distance is inverse to similarities 




```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
#a function to recommend the movies
# first we need to calculate the index of the movie,the compare with each row value and sort the consine distance value then fetch the top 5 movies

def recommend(movie):
    movie_index=new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    # as we sorted the movies consine distance value we lose the index position so enumerate is required to maintain the index position
    #print the movie list
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
